```python
%matplotlib inline
```

```python
from fastai.vision import *
from fastai.callbacks.hooks import *
from fastai.utils.mem import *

from pathlib import Path

path = Path.cwd(); path
```

```python
bs = 64 # batch size
```

```python
data_folder = path/'data'/'genres'

# get images
size=224

tfms = get_transforms(do_flip=False, max_rotate=None, max_warp=None, max_zoom=0)

data = ImageList.from_folder(data_folder, recurse=True).split_by_rand_pct(0.2, seed=2).label_from_folder() \
        .transform(tfms, size=size).databunch()
```

```python
data.show_batch(rows=5, figsize=(9,9))
```

```python
# bn_final = whether to perform batch normalization on the final layer
learn = cnn_learner(data, models.resnet34, metrics=error_rate, bn_final=True)
```

```python
learn.fit_one_cycle(3, slice(1e-1))
```

```python
learn.fit_one_cycle(5, slice(1e-2))
```

```python
learn.fit_one_cycle(3, slice(1e-3))
```

```python
#learn.save('stage-1-36')
learn.load('stage-1-36')
```

```python
learn.lr_find()
learn.recorder.plot()
```

```python
learn.unfreeze()
learn.fit_one_cycle(2, max_lr=slice(1e-6,1e-3))
```
